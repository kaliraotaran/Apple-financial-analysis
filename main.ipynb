```python
# cleaning the data
cleancolumns = df.columns
for col in cleancolumns:
        df[col] = pd.to_numeric(
        df[col].replace({r'\$': '', r',': '', r'%': ''}, regex=True), errors='coerce'
    )
```

```python
df
```

```python
sns.pairplot(data = df)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['year'], df['Revenue (millions)'], label='Revenue (millions)', marker='o')
plt.plot(df['year'], df['Net Income (millions)'], label='Net Income (millions)', marker='o')
plt.title('Revenue and Net Income Trends (2009-2024)')
plt.xlabel('Year')
plt.ylabel('Amount (in millions)')
plt.legend()
plt.show()
```

```python
# Plot Gross Margin and Close Price Trends
fig, ax1 = plt.subplots(figsize=(12, 6))
# Plot Gross Margin
ax1.set_xlabel('Year')
ax1.set_ylabel('Gross Margin (%)', color='tab:blue')
ax1.plot(df['year'], df['Gross Margin'], label='Gross Margin', color='tab:blue', marker='o')
ax1.tick_params(axis='y', labelcolor='tab:blue')
# Overlay Close Price
ax2 = ax1.twinx()
ax2.set_ylabel('Year Close Price', color='tab:green')
ax2.plot(df['year'], df['Year Close Price'], label='Close Price', color='tab:green', linestyle='--', marker='x')
ax2.tick_params(axis='y', labelcolor='tab:green')
fig.tight_layout()
plt.title('Gross Margin and Close Price Trends (2009-2024)')
plt.show()
```

```python
# Calculate Year-over-Year Growth Rate for Revenue and Net Income
df['Revenue Growth (%)'] = df['Revenue (millions)'].pct_change() * 100
df['Net Income Growth (%)'] = df['Net Income (millions)'].pct_change() * 100

# Display Year, Revenue, Net Income, and Growth Rates
growth_columns = ['year', 'Revenue (millions)', 'Revenue Growth (%)', 
                  'Net Income (millions)', 'Net Income Growth (%)']
print(df[growth_columns])
```

```python
# Plot Gross Margin over the years
plt.figure(figsize=(10, 5))
plt.plot(df['year'], df['Gross Margin'], marker='o', color='blue')
plt.title('Gross Margin Over Time')
plt.xlabel('Year')
plt.ylabel('Gross Margin (%)')
plt.grid()
plt.show()
```

```python
# Calculate Debt-to-Assets Ratio
df['Debt-to-Assets Ratio'] = df['Long Term Debt (millions)'] / df['Total Assets (millions)']

# Compare Cash on Hand and Long-Term Debt
financial_health_columns = ['year', 'Cash on Hand (millions)', 
                            'Long Term Debt (millions)', 'Debt-to-Assets Ratio']
print(df[financial_health_columns])

# Plot Cash on Hand vs Debt
plt.figure(figsize=(10, 5))
plt.plot(df['year'], df['Cash on Hand (millions)'], label='Cash on Hand', marker='o')
plt.plot(df['year'], df['Long Term Debt (millions)'], label='Long Term Debt', marker='x')
plt.title('Cash on Hand vs Long Term Debt')
plt.xlabel('Year')
plt.ylabel('Amount (in millions)')
plt.legend()
plt.grid()
plt.show()
```
